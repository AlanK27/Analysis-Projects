```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('SolarPrediction.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['time'] = pd.to_datetime(df['Time'])
df['data'] = pd.to_datetime(df['Data'])

df['hour'] = [n.hour for n in df['time']]
df['dayofweek'] = [n.dayofweek for n in df['data']]
df['month'] = [n.month for n in df['data']]
```

PM/RT = d
pressure Hg  = 0.0334211 atm
Temp F = 273/32 K
R= 0.0821 atm L /K mol
M = 28.97 g/mol

```python
R = 0.0821
M = 28.97
T = 273/32
Hg = 0.0334

df['air_density']=(df['Pressure']/df['Temperature'])*(Hg*M)/(R*T)
```

```python
df.describe()
```

```python
df_h = df.groupby(['hour']).mean().reset_index()
df_m = df.groupby(['month']).mean().reset_index()
df_d = df.groupby(['dayofweek']).mean().reset_index()
```

```python
lis = ['Radiation', 'Temperature', 'Pressure', 'Humidity',
       'WindDirection(Degrees)', 'Speed', 'air_density']
```

```python
lis = ['Radiation', 'Temperature', 'Pressure', 'Humidity',
       'WindDirection(Degrees)', 'Speed', 'air_density']

for n in lis:
    plt.figure()
    pal = sns.color_palette("YlOrRd_r", len(df_h))
    rank = df_h[n].argsort().argsort()
    g = sns.barplot(x='hour', y=n, data= df_h, palette=np.array(pal[::1])[rank])
```

```python
for n in lis:
    plt.figure()
    pal = sns.color_palette("YlOrRd_r", len(df_m))
    rank = df_m[n].argsort().argsort()
    g = sns.barplot(x='month', y=n, data= df_m, palette=np.array(pal[::1])[rank])   
```

```python
for n in lis:
    plt.figure()
    pal = sns.color_palette("YlOrRd_r", len(df_d))
    rank = df_d[n].argsort().argsort()
    g = sns.barplot(x='dayofweek', y=n, data= df_d, palette=np.array(pal[::1])[rank])
```
