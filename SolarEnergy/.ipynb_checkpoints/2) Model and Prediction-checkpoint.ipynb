```python
df = pd.read_csv('SolarPrediction.csv')
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=feats[0], y=feats[1])
plt.xticks(rotation=45)
```

```python
feats
```

```python
X_train_opt = X_train.copy()
futs = feats.copy()
models = []
r2s = []

for i in range(0,6):
    least = np.argmin(futs[1])
    least_important = futs[0][least]
    X_train_opt.drop([least_important], axis=1, inplace=True)
    clf.fit(X_train_opt, y_train)
    feature_importances = clf.feature_importances_
    accuracies = cross_val_score(estimator = clf,
                                 X = X_train_opt,
                                 y = y_train, cv = 3,
                                 scoring = 'r2')
    r2s = np.append(r2s, accuracies.mean())
    models = np.append(models, ", ".join(list(X_train_opt)))
    futs.drop([least], axis=0, inplace=True)
    
feature_selection = pd.DataFrame({'Features':models,'r2 Score':r2s})
feature_selection.head()
```

```python
pd.set_option('display.max_colwidth', 200)
feature_selection
```

```python
feet = feature_selection.iloc[1]['Features']
X = df[feet.split(', ')]
y = df['Radiation']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
clf = RandomForestRegressor()
clf.fit(X_train, y_train)
result = clf.predict(X_test)
```

```python
print(mean_squared_error(y_test, result))
print(mean_absolute_error(y_test, result))
print(r2_score(y_test, result))
print(explained_variance_score(y_test, result))
```

```python
ds = clf.predict(X)

# Create Interactive Plot Using Bokeh
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import Range1d
output_notebook() 

p = figure(plot_width=800, 
           plot_height=400,
           title="Solar Irradiance Predictions",
           x_axis_label='Date and Time',
           y_axis_label='Solar Irradiance',
           x_axis_type="datetime")
p.x_range = Range1d(X.index[0], X.index[600])
p.y_range = Range1d(-100, 1800, bounds=(-100, 1800))
p.line(X.index, y, legend='Observed Solar Irradiance', line_width=1)
p.line(X.index, ds, legend='Predicted Solar Irradiance', line_width=1, line_color="red")

show(p, notebook_handle = True)
```
